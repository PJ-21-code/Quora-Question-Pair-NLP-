```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('/content/train.csv')
```

```python
df.head()
```

```python
df= df.dropna()
df= df.sample(30000)
```

```python
new_df= df[['question1','question2']]
```

```python
new_df.shape
```

```python
questions= list(new_df['question1'])+list(new_df['question2'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer  #array is used in countvectorizer
cv= CountVectorizer(max_features=3000)
```

```python
qarry1,qarry2= np.vsplit(cv.fit_transform(questions).toarray(),2)
```

```python
print(qarry1)
```

```python
temp_df1= pd.DataFrame(qarry1,index=new_df.index)
temp_df2= pd.DataFrame(qarry2,index=new_df.index)
temp_df= pd.concat([temp_df1,temp_df2],axis=1)
temp_df.shape
```

```python
temp_df['is_duplicate']= df['is_duplicate']
```

```python
temp_df.head()
```

```python
x= temp_df.iloc[:,0:-1]
y= temp_df.iloc[:,-1]
```

```python
print(x.sample(10))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
rf= RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
print('accuracy_score= ',accuracy_score(y_pred,y_test))
```
